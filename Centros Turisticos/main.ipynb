```python
# nuevo dataframe seleccionando los valores de Turistas Nacionales Y Extranjeros
# new dataframe selecting the values ​​of National and Foreign Tourists ("Turistas Nacionales Y Extranjeros")

consolidated=df[df["NOM_TIPO"]=="Nacionales Y Extranjeros"]
consolidated.head(5)
```

```python
# dataframe detallado por la columna NOM_TIPO
# dataframe detailed by the "NOM_TIPO" column

detailed=df[df["NOM_TIPO"]!="Nacionales Y Extranjeros"]
detailed.head(5)
```

```python
# Numero de visitantes por año 
# Number of visitors per year

column="ANIO"
visits = consolidated.groupby(column)['TOTAL'].sum()

x=visits.index
y=visits

plt.figure(figsize=(15, 6))
plt.bar(x, y)

plt.xlabel("Años")
plt.ylabel("# de Visitantes")
plt.title(f"Registros historicos de visitantes por año")
plt.xticks(x, rotation=90)
plt.show()
```

```python
# sumatoria de visitantes por año
# sum of visitors per year

visits
```

```python
# sumatoria total visitantes del df
# total sum of visitors from the city

visits.sum()
```

```python
#numero de visitantes por mes de un año determinado 
#number of visitors per month of a given year

year=2023
partial_data=consolidated[consolidated['ANIO'] == year]

column="COD_MES"
visits = partial_data.groupby(column)['TOTAL'].sum()

x=visits.index
y=visits

months = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']


plt.figure(figsize=(15, 6))
plt.plot(months, y, marker='o', linestyle='-')

plt.xlabel(f"{column}")
plt.ylabel("# de Visitantes")
plt.title(f"Registros visitantes correspondiente al año {year}")

plt.xticks(months, rotation=45)

plt.grid()
plt.show()

```

```python
# numero de turistas por mes
# number of visits per month

visits
```

```python
#numero de visitantes por museo de un año determinado 
#number of visitors per museum in a given year

year=2023
partial_data=consolidated[consolidated['ANIO'] == year]

column="NOM_MUSEO"
visits = partial_data.groupby(column)['TOTAL'].sum()

x=visits.index
y=visits

plt.figure(figsize=(10, 6))
plt.bar(x, y, linestyle='-')

plt.xlabel(f"{column}")
plt.ylabel("# de Visitantes")
plt.title(f"Visitas a museos de sitio correspondientes al año {year}")

plt.xticks(x, rotation=90)

plt.show()
```

```python
# numero de turistas por año y museo
# number of tourists per year and museum

visits
```

```python
#numero de visitantes por ciudad de un año determinado 
#number of visitors per city in a given year

year=2023
partial_data=consolidated[consolidated['ANIO'] == year]

column="NOM_DPTO"
visits = partial_data.groupby(column)['TOTAL'].sum()

x=visits.index
y=visits

plt.figure(figsize=(15, 6))
plt.bar(x, y, linestyle='-')

plt.xlabel(f"{column}")
plt.ylabel("# de Visitantes")
plt.title(f"Registros correspondiente al año {year}")

plt.xticks(x, rotation=90)

plt.show()

```

```python
# numero de turistas por año determinado y departamento
# number of tourists per given year and department

visits
```

```python
# numero de turistas por procedencia
# number of tourists by origin

year=2023
partial_data=detailed[detailed['ANIO'] == year]
column="NOM_TIPO"
visits = partial_data.groupby(column)['TOTAL'].sum()

x=visits.index
y=visits

plt.figure(figsize=(15, 6))
plt.bar(x, y)

plt.xlabel("Año")
plt.ylabel("Frecuencia")
plt.title(f"Registros correspondiente al año {year}")

plt.xticks(x, rotation=90)

plt.show()
```
