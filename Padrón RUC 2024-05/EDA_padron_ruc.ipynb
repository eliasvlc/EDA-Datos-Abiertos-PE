<p> En

Desde el estado y la condición de las empresas hasta su actividad económica principal y secundaria, pasando por

<h3>Descripción del Dataset</h3>
    <ul>
        <li><strong>RUC:</strong> Número de Registro Único de Contribuyente, identificación única asignada a cada entidad legal en el Perú para fines tributarios y administrativos.</li>
        <li><strong>Estado:</strong> Estado actual de la entidad, que puede ser "ACTIVO", "BAJA PROVISIONAL POR OFICIO" u otro estado que indique su situación legal o administrativa.</li>
        <li><strong>Condicion:</strong> Condición actual de la entidad, que puede indicar si está "HABIDO" o "NO HABIDO", refiriéndose a si la entidad está localizable o no.</li>
        <li><strong>Tipo:</strong> Tipo de entidad, que puede variar entre diferentes formas legales de negocio, como "PERSONA NATURAL SIN EMPRESA", "SOCIEDAD ANONIMA CERRADA", entre otros.</li>
        <li><strong>Actividad_Economica_CIIU_revision3_Principal:</strong> Principal actividad económica de la entidad según la Clasificación Industrial Internacional Uniforme (CIIU) revisión 3.</li>
        <li><strong>Actividad_Economica_CIIU_revision3_Secundaria:</strong> Actividad económica secundaria de la entidad según la Clasificación Industrial Internacional Uniforme (CIIU) revisión 3.</li>
        <li><strong>Actividad_Economica_CIIU_revision4_Principal:</strong> Principal actividad económica de la entidad según la Clasificación Industrial Internacional Uniforme (CIIU) revisión 4.</li>
        <li><strong>NroTrab:</strong> Número de trabajadores empleados por la entidad.</li>
        <li><strong>TipoFacturacion:</strong> Método utilizado por la entidad para la facturación, que puede ser "MANUAL", "COMPUTARIZADO", "ELECTRONICO", entre otros.</li>
        <li><strong>TipoContabilidad:</strong> Tipo de contabilidad utilizado por la entidad, que puede ser "MANUAL", "COMPUTARIZADO", "SIN SISTEMA", entre otros.</li>
        <li><strong>ComercioExterior:</strong> Indicador de la participación de la entidad en el comercio exterior, que puede ser "SIN ACTIVIDAD", "IMPORTADOR/EXPORTADOR", "IMPORTADOR", "EXPORTADOR", entre otros.</li>
        <li><strong>UBIGEO:</strong> Código de Ubicación Geográfica, que identifica la ubicación geográfica de la entidad según el sistema de codificación utilizado en el Perú.</li>
        <li><strong>Departamento:</strong> Departamento geográfico donde se encuentra la entidad.</li>
        <li><strong>Provincia:</strong> Provincia geográfica donde se encuentra la entidad.</li>
        <li><strong>Distrito:</strong> Distrito geográfico donde se encuentra la entidad.</li>
        <li><strong>PERIODO_PUBLICACION:</strong> Fecha o periodo de publicación de los datos.</li>
    </ul>
